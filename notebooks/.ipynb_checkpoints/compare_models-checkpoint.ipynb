```python
import os 

import numpy as np
import scipy.stats as sps
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random
from matplotlib.cm import get_cmap
from tqdm import tqdm

import sys, os
sys.path += [os.path.abspath(os.pardir + '/src')]
print(sys.path)

from experiment import init_random_state, BanditLoopExperiment, get_ts_model, get_random_model, get_epsilon_greedy_model, get_optimal_model

sns.set(font_scale=2.5, palette='tab20', font="Times New Roman") 
```

```python
seed = 42

p = 1.
Q = 1
ws = [4]
b = 0.0

T = 5000
M = 10
l = 4
```

```python
interests, _, responses = [], [], []
interests_TS, TS_paramss_TS, responses_TS = [], [], []
interests_opt, TS_paramss_opt, responses_opt = [], [], []
interests_eps, TS_paramss_eps, responses_eps = [], [], []
```

```python
%%time
for w in tqdm(ws):
    init_random_state(seed)
    
    bandit = lambda: get_random_model(M=M, l=l)
    exp = BanditLoopExperiment(bandit, "random bandit")
    exp.prepare(w=w, Q=Q, p=p, b=b)
    exp.run_experiment(T=T)
    
    results = exp.get_as_np()
    interests.append(results.interest)
    _.append(results.TS_params)
    responses.append(results.response)
    
for w in tqdm(ws):
    init_random_state(seed)
    
    bandit = lambda: get_ts_model(M=M, l=l)
    exp = BanditLoopExperiment(bandit, "TS bandit")
    exp.prepare(w=w, Q=Q, p=p, b=b)
    exp.run_experiment(T=T)
    
    results = exp.get_as_np()
    interests_TS.append(results.interest)
    TS_paramss_TS.append(results.TS_params)
    responses_TS.append(results.response)

for w in tqdm(ws):
    init_random_state(seed)
    
    bandit = lambda: get_epsilon_greedy_model(M=M, l=l)
    exp = BanditLoopExperiment(bandit, "greedy bandit")
    exp.prepare(w=w, Q=Q, p=p, b=b)
    exp.run_experiment(T=T)
    
    results = exp.get_as_np()
    interests_eps.append(results.interest)
    _.append(results.TS_params)
    responses_eps.append(results.response)

for w in tqdm(ws):
    init_random_state(seed)
    
    bandit = lambda: get_optimal_model(M=M, l=l)
    exp = BanditLoopExperiment(bandit, "optimal bandit")
    exp.prepare(w=w, Q=Q, p=p, b=b)
    exp.run_experiment(T=T)
    
    results = exp.get_as_np()
    interests_opt.append(results.interest)
    _.append(results.TS_params)
    responses_opt.append(results.response)
```

```python
plt.figure(figsize=(16, 10))
for i, w in enumerate(ws):
    plt.plot(np.linalg.norm(interests[i] - interests[i][0], axis=1)**2, linewidth = 3, label=f'w = {round(w,3)}, random')
for i, w in enumerate(ws):
    plt.plot(np.linalg.norm(interests_TS[i] - interests_TS[i][0], axis=1)**2, linewidth = 3,label=f'w = {round(w,3)}, TS')
for i, w in enumerate(ws):
    plt.plot(np.linalg.norm(interests_opt[i] - interests_opt[i][0], axis=1)**2, linewidth = 3, label=f'w = {round(w,3)}, Optimal')
for i, w in enumerate(ws):
    plt.plot(np.linalg.norm(interests_eps[i] - interests_eps[i][0], axis=1)**2, linewidth = 3, label=f'w = {round(w,3)}, Eps')

plt.yscale('log')
plt.ylabel(r'$\|\mu_t - \mu_0 \|^2$')
plt.xlabel('Шаг')
plt.legend()
plt.savefig('../figures/compare_models_w.pdf')
```

```python
seed = 42

p = 1.
Q = 1
w = 0
bs = [0.7]

T = 5000
M = 10
l = 4

interests, _, responses = [], [], []
interests_TS, TS_paramss_TS, responses_TS = [], [], []
interests_opt, TS_paramss_opt, responses_opt = [], [], []
interests_eps, TS_paramss_eps, responses_eps = [], [], []

for b in tqdm(bs):
    init_random_state(seed)
    
    bandit = lambda: get_random_model(M=M, l=l)
    exp = BanditLoopExperiment(bandit, "random bandit")
    exp.prepare(w=w, Q=Q, p=p, b=b)
    exp.run_experiment(T=T)
    
    results = exp.get_as_np()
    interests.append(results.interest)
    _.append(results.TS_params)
    responses.append(results.response)
    
for b in tqdm(bs):
    init_random_state(seed)
    
    bandit = lambda: get_ts_model(M=M, l=l)
    exp = BanditLoopExperiment(bandit, "TS bandit")
    exp.prepare(w=w, Q=Q, p=p, b=b)
    exp.run_experiment(T=T)
    
    results = exp.get_as_np()
    interests_TS.append(results.interest)
    TS_paramss_TS.append(results.TS_params)
    responses_TS.append(results.response)

for b in tqdm(bs):
    init_random_state(seed)
    
    bandit = lambda: get_epsilon_greedy_model(M=M, l=l)
    exp = BanditLoopExperiment(bandit, "greedy bandit")
    exp.prepare(w=w, Q=Q, p=p, b=b)
    exp.run_experiment(T=T)
    
    results = exp.get_as_np()
    interests_eps.append(results.interest)
    _.append(results.TS_params)
    responses_eps.append(results.response)

for b in tqdm(bs):
    init_random_state(seed)
    
    bandit = lambda: get_optimal_model(M=M, l=l)
    exp = BanditLoopExperiment(bandit, "optimal bandit")
    exp.prepare(w=w, Q=Q, p=p, b=b)
    exp.run_experiment(T=T)
    
    results = exp.get_as_np()
    interests_opt.append(results.interest)
    _.append(results.TS_params)
    responses_opt.append(results.response)
    
plt.figure(figsize=(16, 10))
for i, b in enumerate(bs):
    plt.plot(np.linalg.norm(interests[i] - interests[i][0], axis=1)**2, linewidth = 3, label=f'b = {round(b,3)}, random')
for i, b in enumerate(bs):
    plt.plot(np.linalg.norm(interests_TS[i] - interests_TS[i][0], axis=1)**2, linewidth = 3,label=f'b = {round(b,3)}, TS')
for i, b in enumerate(bs):
    plt.plot(np.linalg.norm(interests_opt[i] - interests_opt[i][0], axis=1)**2, linewidth = 3, label=f'b = {round(b,3)}, Optimal')
for i, b in enumerate(bs):
    plt.plot(np.linalg.norm(interests_eps[i] - interests_eps[i][0], axis=1)**2, linewidth = 3, label=f'b = {round(b,3)}, Eps')

plt.yscale('log')
plt.ylabel(r'$\|\mu_t - \mu_0 \|^2$')
plt.xlabel('Шаг')
plt.legend()
plt.savefig('../figures/compare_model_b.pdf')
```
